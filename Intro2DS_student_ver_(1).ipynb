```
# TODO: Plot your hypothesis as per the plot above.
def give_code_explanation(code):
  global df
  code = code.replace("_",'.')
  return df[df.Indicator_Code == code].Indicator_Name.values[0]

def table_code_explanation(numeric_df):
  pd.set_option('max_colwidth', 800)
  codes = numeric_df.columns
  explanations = [ give_code_explanation(x) for x in codes ]
  explanations_table = pd.DataFrame({
      'Code' : codes,
      'Explanation' : explanations
  })
  return explanations_table

table_code_explanation(numeric_df)
```
